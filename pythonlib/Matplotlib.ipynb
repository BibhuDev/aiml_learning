Matplotlib is used for making plots

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x= np.linspace(0,10,100) ##prints 100 values between 0 and 10, equally distant from each other
```

```python
y= np.sin(x)
z= np.cos(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
x= np.linspace(0,6.2831,500)
y= np.sin(x)
z= np.abs(y/5)

plt.xlabel("angle")
plt.ylabel("absolute sine curve")


plt.plot(x,z)
plt.show()

```

```python
plt.plot(x, np.sin(x), 'r')
plt.plot(x, np.cos(x), 'g--')
plt.show()
```

Bar plot

```python
figure= plt.figure()
ax= figure.add_axes([0,0,1,1])
lang=['hindi','odia','english','japanese','korean']
people=[122, 75, 189, 45, 2]
ax.bar(lang,people)
plt.xlabel("Languages")
plt.ylabel("Number of people")
plt.title("Popularity of languages")
plt.show()
```

pie chart

```python
figure= plt.figure()
ax= figure.add_axes([0,0,1,1])
lang=['hindi','odia','english','japanese','korean']
people=[122, 75, 189, 45, 2]
ax.pie(people, labels= lang, autopct='%1.1f%%')
plt.show()
```

Scatter plot

```python
fig= plt.figure()
ax= fig.add_axes([0,0,1,1])
x=[100,30,56,22,44,68,91]
y=[21,32,43,54,65,76,87]
plt.xlabel("weight")
plt.ylabel("age")
ax.scatter(x,y)
plt.show()
```

3D plot

```python
fig= plt.figure()
ax= plt.axes(projection='3d')
z= np.linspace(0,15,1000)
x= np.sin(z)
y= np.cos(z)
ax.plot3D(x,y,z, 'g--')
plt.show()
```
